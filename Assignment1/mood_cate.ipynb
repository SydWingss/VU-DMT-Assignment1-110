```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('../Assignment1/divide_by_variable/mood.csv')
```

```python
# Plot the distribution of 'mood'
sns.histplot(df['mood'], kde=True)
plt.title('Distribution of Mood')
plt.show()

# Define the boundaries for low, medium and high mood
low_boundary = 6
high_boundary = 8

# Create a new column 'mood_level'
df['mood_level'] = 'low'
df.loc[df['mood'] > low_boundary, 'mood_level'] = 'medium'
df.loc[df['mood'] > high_boundary, 'mood_level'] = 'high'
```

```python
df['mood_level'].value_counts()
```

```python
df.to_csv('mood_level.csv', index=False)
```
